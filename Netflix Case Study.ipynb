# How has the number of movies released per year changed over the last 20-30 years?

```python
a=netflix.loc[(netflix["type"]=="Movie")&(netflix["release_year"]>1990)].groupby("release_year")["show_id"].count()
```

```python
plt.figure(figsize=(20,10))
plt.bar(a.index,a)
plt.xticks(rotation =45, fontsize=10)
plt.show()
```

```python
# Inference and Recommendation:
#      Netflix has increased the number of movies per year since last 30 years for almost every year.
#      The relative increase in movies per year shows Netflix is on right path and is coming up with more quality content for its users.
```

# What is the best time to launch a TV show?

```python
netflix.head()
```

```python
c=netflix[netflix["type"]=="TV Show"]
```

```python
new=c["date_added"].str.split(n=1,expand=True)
```

```python
c["month"]=new[0]
```

```python
c.head()
```

```python
c.groupby("month")["show_id"].count().sort_values(ascending=False)
```

```python
d=c.groupby("month")["show_id"].count().sort_values(ascending=False)
d_ind=c.groupby("month")["show_id"].count().sort_values(ascending=False).index
```

```python
plt.figure(figsize=(20,10))
plt.bar(d_ind,d)
plt.xlabel("Months")
plt.ylabel("Counts")
plt.xticks(rotation=45,fontsize=10)
plt.title("TV Show release month")
plt.show()
```

```python
#Inference:
# The data shows that the maximum TV shows are released in months of December and July in Netflix.

#Recommendation:
# According the data , it is advisable to release new show during the month of December or July.
```

# Does Netflix has more focus on TV Shows than movies in recent years

```python
netflix["year"]=netflix["date_added"].str.split(expand=True)[2]
```

```python
netflix.head()
```

```python
netflix["year"]=netflix["year"].astype(int)
```

```python
netflix.loc[netflix["year"]>2015].groupby(["year","type"])["show_id"].count()
```

```python
g=netflix.loc[netflix["year"]>2015].groupby(["year","type"])["show_id"].count().reset_index()
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=g.loc[g["type"]=="Movie"],x='year',y="show_id",label="Movie",color="red")
sns.lineplot(data=g.loc[g["type"]=="TV Show"],x='year',y="show_id",label="TV Show",color="green")

plt.legend()
plt.grid()
plt.show()
```

```python
#Inference:It is not the case that the Netflix focus has been towards TV Shows lately rather it is other way
#Recommendation: It would be advisable that the focus is provided on both the types of contents.
```

# Duration based analysis


```python
netflix.head()
```

```python
netflix["eff_duration"]=netflix["duration"].str.split(expand=True)[0]
```

```python
netflix["eff_duration"].head()
```

```python
#For movies 
```

```python
netflix["eff_duration"].astype(int)
```

```python
netflix[netflix["type"]=="Movie"].groupby("year")["eff_duration"].median().reset_index()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=netflix[netflix["type"]=="Movie"].groupby("year")["eff_duration"].median().reset_index(),x="year",y="eff_duration")
plt.legend(labels="Movie",fontsize=12)
plt.ylabel("eff_duration in mins" , fontsize=14)
plt.xlabel("year",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=netflix[netflix["type"]=="TV Show"].groupby("year")["eff_duration"].median().reset_index(),x="year",y="eff_duration")
plt.legend(labels="TV Show",fontsize=12)
plt.ylabel("eff_duration in seasons" , fontsize=14)
plt.xlabel("year",fontsize=14)
plt.show()
```
